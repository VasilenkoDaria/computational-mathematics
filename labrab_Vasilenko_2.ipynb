### Функции Гаусса, Зубец и их производные

```python
def gauss(x):
    Dt = 0.4
    return 1./(np.pi*Dt)**0.5*np.exp(-x*x/Dt)

def gauss_deriv(x):
    Dt = 0.4
    return -2*x/Dt/(np.pi*Dt)**0.5*np.exp(-x*x/Dt)
```

```python
def prong(x):
    if x < -0.5 or x > 0.5:
        return 0.
    if x >= -0.5 and x < 0:
        return x + 0.5
    if x <= 0.5 and x >= 0:
        return 0.5 - x
    
def prong_deriv(x):
    if x < -0.5 or x > 0.5:
        return 0.
    if x >= -0.5 and x < 0:
        return 1
    if x <= 0.5 and x >= 0:
        return -1 
```

### Нормы

```python
from scipy.spatial import distance

def L1_norm(x, h):
    return sum(abs(x) * h)

def L2_norm(x, h):
    return np.linalg.norm(x) * h ** 0.5

def Linf_norm(x):
    return max(abs(x))
```

### Вычисление градиентов в центре ячейки

```python
def central_3(x, f): 
    h = x[1:] - x[:-1]       
    return ((f(x[1]) - f(x[0]))*h[1]**2 + (f(x[2]) - f(x[1]))*h[0]**2)/(h[0]*h[1]**2 + h[1]*h[0]**2)


def central_35(x, f): 
    return central_3(x[1:-1], f)*4/3 - central_3(x[::2], f)/3


def central_5(x, f): 
    grad = 0.   
    for i in [0, 1, 3, 4]:
        P = 1.      
        for k in [0, 1, 3, 4]:
            if k == i: continue
            P *= (x[2] - x[k])/(x[i] - x[k])          
        grad += P*(f(x[i]) - f(x[2]))/(x[i] - x[2])
    return grad
```

Доп. комментарии: На неравномерной сетке, используя трёхточечный шаблон, можно получить второй порядок аппроксимации производной(с3). Для повышения точности метода рассмотрим формулу 4-го порядка аппроксимации на пятиточечном шаблоне в случае равномерной сетки, её можно переписать как комбинацию двух центральных разностей – на узком и широком шаблонах. Полученная схема будет иметь 4-го порядок точности на равномерной сетке. На неравномерной сетке порядок такого подхода будет снижаться до 2-го, но все равно можно ожидать уточнения по сравнению с исходным методом.(с35) Используя 5-точечный шаблон, можно построить и более точную аппроксимацию производной: формула Лагранжа(с5).

### Вычисление касательных градиентов на грани ячейки

```python
def tangential_2(x, df, f): 
    h = (x[1:] - x[:-1])
    return (df(x[0])*h[1] + df(x[2])*h[0])/(h[0] + h[1])


def tangential_24(x, df, f): 
    return tangential_2(x[1:-1], df, f)*9/8 - tangential_2(x[::2], df, f)/8


def tangential_4(x, df, f): 
    h = x[1:] - x[:-1]
    grad = 0.  
    for i in [0, 1, 3, 4]:
        P = 1.     
        for k in [0, 1, 3, 4]:
            if k == i: continue
            P *= (x[2] - x[k])/(x[i] - x[k])
        grad += P*df(x[i])     
    return grad
```

Доп. комментарии: Вторая задача состоит в интерполяции ранее вычисленных производных из центров ячеек на грань. В качестве базового подхода возьмем линейную интерполяцию по 2 ближайшим ячейкам. Она имеет второй порядок аппроксимации(t_2) Идея повышения точности аналогична предыдущей задаче. Возьмем формулу 4-точечной интерполяции имеющую 4-й порядок точности на равномерном шаблоне и представим ее как комбинацию полусумм производных на узком и широком шаблонах. Формально заменим полусуммы на линейные интерполяции и снова получим метод, обладающий 4-м порядком точности на равномерной сетке и 2 на неравномерной.(t_24) Дальнейшее уточнение можно сделать, выписав интерполяционную формулу Лагранжа, которая будет обладать 4-м порядком точности на неравномерном шаблоне (t_4)

### Вычисление нормальных градиентов на грани ячейки

```python
def normal_2(x, df, f):
    h = x[1:] - x[:-1]     
    return (f(x[2]) - f(x[0]))/(h[0] + h[1]) - (h[1] - h[0])*(df(x[2]) - df(x[0]))/(h[0] + h[1])/2
    #return (f(x[2]) - f(x[0]))/(h[0] + h[1]) - (h[1] - h[0])*(df(x[2]) - df(x[0]))/(h[0] + h[1])  

       
def normal_24(x, df, f): 
    return normal_2(x[1:-1], df, f)*9/8 - normal_2(x[::2], df, f)/8


def normal_4(x, df, f):
    h = x[1:] - x[:-1]
    grad = 0.   
    for i in [0, 1, 3, 4]:
        S = 0.
        for k in [0, 1, 3, 4]:
            if k == i: continue
            P = 1.  
            for q in [0, 1, 3, 4]:
                if q == i or q == k: continue
                P *= (x[2] - x[q])
            S += P
        for k in [0, 1, 3, 4]:
            if k == i: continue
            S /= (x[i] - x[k])
        grad += S*f(x[i])
    
    return grad
```

Доп. комментарии:  Имея значения функции и ее производных в центрах ячеек, необходимо восстановить значение производной на грани в направлении нормали к ней. Предлагается вычислять производную на грани по двум ближайшим точкам, а затем в случае неравномерной сетки корректировать ее с помощью производных из этих же двух точек. Так как интерполяция, используемая для касательных градиентов, имеет низкое качество, так как приводит к потере взаимодействия между четными и нечетными слоями ячеек. Этот подход обеспечивает второй порядок точности на неравномерных сетках.(n_2) Точность аппроксимации можно повысить, взяв производные на узком и широком шаблонах. Полученный метод снова будет обладать 4-м порядком точности на равномерной сетке и 2-м – на неравномерной. (n_24) Продифференцировав интерполяционную формулу Лагранжа, можно выразить производную на грани через значения функции в центрах ячеек(  получим формулу 4-го порядка точности на равномерной сетке и 3-го – на неравномерной. ) (n_4) 

```python
N = 1000
x = np.linspace(-2, 2, N + 1)
h1 = 0.1
h2 = 0.05

h = [-0.2, 0., 0.1, 0.2, 0.4]
nl_h = [-0.075, 0., 0.05, 0.25, 0.85]

norm_2 = np.zeros_like(x)
norm_4 = np.copy(norm_2)
norm_24 = np.copy(norm_2)

norm_2_nl = np.copy(norm_2)
norm_4_nl = np.copy(norm_2)
norm_24_nl = np.copy(norm_2)

for i in range(N + 1):
    norm_2[i] = normal_2(h[1:-1] + x[i], gauss_deriv, gauss)
    norm_4[i] = normal_4(h + x[i], gauss_deriv, gauss)
    norm_24[i] = normal_24(h + x[i], gauss_deriv, gauss)
    
for i in range(N + 1):
    norm_2_nl[i] = normal_2(nl_h[1:-1] + x[i], gauss_deriv, gauss)
    norm_4_nl[i] = normal_4(nl_h + x[i], gauss_deriv, gauss)
    norm_24_nl[i] = normal_24(nl_h + x[i], gauss_deriv, gauss)
```
